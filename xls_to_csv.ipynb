```python
import pandas as pd
import os
```

```python

def excel_to_csv(file,to_file):
    data_xls=pd.read_excel(file,sheet_name=0)
    data_xls.to_csv(to_file,encoding='utf_8_sig')
 
  
  # 读取一个目录里面的所有文件：
def read_path(path):
    dirs=os.listdir(path)
    return dirs

```

```python
#主函数
def main():
    #源文件路径
    source = "C:\\Users\\hongs\\zerenhongshen\\CIED(1998-2013)"
 
    #目标文件路径
    ob = "C:\\Users\\hongs\\zerenhongshen\\CIED(1998-2013)2"
 
    #将源文件路径里面的文件转换成列表file_list
    file_list=[source+"\\"+i for i in read_path(source)]
    j = 0
    #建立循环对于每个文件调用excel_to_csv()
    for it in file_list:
        #给目标文件新建一些名字列表
        j_mid = read_path(source)[j]
        j_csv = ob+"\\"+j_mid+".csv"
        excel_to_csv(it,j_csv)
        print(it)
        j=j+1
 
if __name__ == '__main__':

    main()
```
